# Association Rule Mining:
•	Implement an Apriori algorithm using tool like python with libraries such as Pandas and Mlxtend etc.
•	 Apply association rule mining techniques to the pre-processed dataset to discover interesting relationships between products purchased together.

```python
#checking the shape of data
df.shape 
```

```python
#to find the frequent itemsets in dataset

from mlxtend.frequent_patterns import apriori,association_rules

frequent_items=apriori(df,min_support=0.01,use_colnames=True)  #based on the value of min_support the number of rules will depend
frequent_items             #the min_support  ensures that only item sets appering  in atleat 0.4% of transcation are consider
```

```python
#finding the assocaion rules 
                         #frequent items found in previous
rules=association_rules(frequent_items,metric='lift',min_threshold=4,num_itemsets=20)
                                     #we are using the lift ratio  min threshold=1 i.e lift ratio
```

```python
rules
```

```python
#shorting the lift values in decending order

rules['lift'].sort_values(ascending=False)
```

```python
#from the above lift values we can say that

#12.84 → Very strong positive association. The presence of the antecedent makes the consequent 12.84 times more likely than random chance. This suggests a highly significant relationship.
# 8.22 → Strong association, meaning the items are frequently purchased or appear together much more than expected.
# 7.59 → Still a strong lift, showing a meaningful connection between items.
# 6.47 → Moderate to strong association, indicating a clear pattern but slightly less impactful.
#4.37 → Moderate association, meaning the items are 4.37 times more likely to appear together than by random occurrence.
```

```python
#antecedents=customer come to buy
#consequence=customer who als buy these with antecedent
#support=probabilty of buying
#confidence=Condition probability to buy
#lift=calculates the "strength of rule" if {lift>1=positive association},{lift value=1 no relation},{lift value< 1 negative association}
#support measures how frequently an itemset appers in a dataset how much higher- (>0.05 or 5%)  the value means the itemset is commonly occured
#confidence measures how often item B appers in transcation that contains item A - Confidence > 50% is generally considers strong
#if support is low(0.01) and confidence is high(0.50)then it is a fraud transcation
 
```

```python
#sorting the dataset on the lift values as decending

rules.sort_values('lift',ascending=False)
```

```python
#With "min_support=0.01",and "min_lift=4" we are forming 9 strong association rules
```

# Method 2 using apyori library

```python
#without data transformation

df_1=pd.read_excel("Online retail.xlsx",header=None)
df_1
```

```python
#apriori algorithm  will generate the association rule whenthe data is in list form

trans=[]
for i in range(0,7501):  # for rows there are 7501 rows 
    trans.append([str(df_1.values[i,j]) for j in range(0,20)])  #j loop is for columns
```

```python
print(trans[0])
```

```python
from apyori import apriori

results=apriori(transactions=trans,  #list of data
                min_support=0.01,
                min_confidence=0.2,
                min_lift=3,
                min_length=2,       #conbination of transcation
                max_length=3)

results  # generate a generator in opps
```

```python
#converting to list

report=list(results)
report
```

```python
# Extracting association rules in a structured format
data = []

for item in report:
    support = item.support
    for stat in item.ordered_statistics:
        base = tuple(stat.items_base)
        additem = tuple(stat.items_add)
        confidence = stat.confidence
        lift = stat.lift

        data.append([base, additem, support, confidence, lift])

# Creating the DataFrame
df = pd.DataFrame(data, columns=['Base Item/Anticend', 'Add Item/consequence', 'Support', 'Confidence', 'Lift'])

# Display the DataFrame
print(df)
```
