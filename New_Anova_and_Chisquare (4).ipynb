# **Chisquare**

```python
import pandas as pd
import numpy as np


df = pd.read_csv("mtcars.csv")
```

```python
df.head()
```

```python
crosstable = pd.crosstab(df['cyl'],df['gear'])
```

```python
crosstable
```

```python
from scipy.stats import chi2_contingency
chi, p, dof, expected = chi2_contingency(crosstable)
```

```python

print(chi,p,dof)
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(crosstable)
```

```python
#Since p value is less than alpha, we reject Ho, that is accept H1.
#Hence we conclude that  both cyl and gear are dependent variables
```
