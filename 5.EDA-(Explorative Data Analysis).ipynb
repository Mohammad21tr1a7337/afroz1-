```python
# Define colors for different elements in boxplot
box_color = "blue"      #the rectangle box color
median_color = "red"    #for the middle line in box /median line
whisker_color = "green"  #the verticalline that join the upper bound
cap_color = "purple"     #the upper/lowerbound line
flier_color = "black"     #otlayer points color

# Loop through each column in the DataFrame
for column in df.columns:
    plt.boxplot(df[column],             #generate bocplot for each columns
                boxprops=dict(color=box_color),  #boxprops is the boxplot propertis
                medianprops=dict(color=median_color, linewidth=2),   #median line propertis
                whiskerprops=dict(color=whisker_color, linestyle="--"),  #the whister line the vert one with line style
                capprops=dict(color=cap_color),                        #the upp/lower bound line colors
                flierprops=dict(marker="o", color=flier_color, markersize=8))
    
    plt.title(column, fontsize=14, color="darkblue")  # Change title color
    plt.show()  # Display the boxplot
    print()  # Gives a gap after every plot
    
```
